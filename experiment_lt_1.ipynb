```python
import pandas as pd
import numpy as np
import bokeh
bokeh.io.output_notebook()
```

```python
data = pd.read_csv('/Users/jason.katz/Downloads/search_results1.csv')
```

```python
data['lt'] = 1
for index, row in data.iterrows():
    data['lt'][index] = row['job_id'][-2:]
```

```python
data['game'] = 'id'
for index, row in data.iterrows():
    data['game'][index] = row['job_id'][14:20]
```

```python
data[data['game'] == 'ER0429']
```

```python
scat1 = bokeh.charts.Scatter(data, x='lt', y='mota', color='game')
bokeh.io.show(scat1)
```

```python
data_lt = data.copy()
data_lt['loss'] = data_lt['lt']
data_lt = data_lt.groupby(['lt']).mean()
data_lt
```

```python
scat2 = bokeh.charts.Scatter(data_lt, x = 'loss', y='mota')
scat3 = bokeh.charts.Scatter(data_lt, x = 'loss', y='motp')
scat4 = bokeh.charts.Scatter(data_lt, x = 'loss', y='mostly_tracked')
scat5 = bokeh.charts.Scatter(data_lt, x = 'loss', y='mt_auc')
bokeh.io.show(bokeh.layouts.column(bokeh.layouts.row(scat2, scat3), bokeh.layouts.row(scat4, scat5)))
```
