```python
import pandas as pd
import numpy as np
from bokeh.charts import Scatter
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
```

```python
output_notebook()
```

```python
data = pd.read_csv('/Users/jason.katz/Downloads/search_results5.csv')
```

```python
data['lt'] = 1
for index, row in data.iterrows():
    data['lt'][index] = row['job_id'][-2:]
```

```python
data['game'] = 'id'
for index, row in data.iterrows():
    data['game'][index] = row['job_id'][18:24]
```

```python
data = data[data['lt']!=24]
data
```

```python
scat1 = Scatter(data, x='lt', y='mota', color='game')
show(scat1)
```

```python
data_lt = data.copy()
data_lt['loss'] = data_lt['lt']
data_lt = data_lt.groupby(['lt']).mean()
data_lt
```

```python
scat2 = Scatter(data_lt, x = 'loss', y='mota')
scat3 = Scatter(data_lt, x = 'loss', y='motp')
scat4 = Scatter(data_lt, x = 'loss', y='mostly_tracked')
scat5 = Scatter(data_lt, x = 'loss', y='mt_auc')
show(column(row(scat2, scat3), row(scat4, scat5)))
```

```python
data['mota'].groupby(data['lt']).describe()
```

```python
scat6 = Scatter(data_lt, x = 'loss', y='mota');scat6.y_range.start,scat6.y_range.end = (0,1)
scat7 = Scatter(data_lt, x = 'loss', y='motp');scat7.y_range.start,scat7.y_range.end = (0,1)
scat8 = Scatter(data_lt, x = 'loss', y='mostly_tracked');scat8.y_range.start,scat8.y_range.end = (0,1)
scat9 = Scatter(data_lt, x = 'loss', y='mt_auc');scat9.y_range.start,scat9.y_range.end = (0,1)
show(column(row(scat6, scat7), row(scat8, scat9)))
```
