```python
import pandas as pd
import numpy as np
from bokeh.charts import Scatter
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
```

```python
output_notebook()
```

```python
data = pd.read_csv('/Users/jason.katz/Downloads/hung1.csv')
```

```python
ind = [index for index,row in data.iterrows() if 'HUNG-1' in row['job_id']]
data = data.loc[ind]
```

```python
data['ht'] = 1
for index, row in data.iterrows():
    data['ht'][index] = row['job_id'][-3:]
data['ht'][data['ht']<0]=data['ht'][data['ht']<0]*-1
```

```python
data['game'] = 'id'
for index, row in data.iterrows():
    data['game'][index] = row['job_id'][20:26]
```

```python
scat1 = Scatter(data, x='ht', y='mota', color='game')
show(scat1)
```

```python
data_ht = data.copy()
data_ht['hungarian'] = data_ht['ht']
data_ht = data_ht.groupby(['ht']).mean()
data_ht
```

```python
scat2 = Scatter(data_ht, x = 'hungarian', y='mota')
scat3 = Scatter(data_ht, x = 'hungarian', y='motp')
scat4 = Scatter(data_ht, x = 'hungarian', y='mostly_tracked')
scat5 = Scatter(data_ht, x = 'hungarian', y='mt_auc')
from bokeh.layouts import column, row
show(column(row(scat2, scat3), row(scat4, scat5)))
```
