```python
import pandas as pd
import numpy as np
from bokeh.charts import Scatter
from bokeh.io import show
bokeh.io.output_notebook()
```

```python
data = pd.read_csv('/Users/jason.katz/Downloads/search_results4.csv')
```

```python
data['lt'] = 1
for index, row in data.iterrows():
    data['lt'][index] = row['job_id'][-2:]
```

```python
data['game'] = 'id'
for index, row in data.iterrows():
    data['game'][index] = row['job_id'][16:22]
```

```python
data = data[data['lt']!=24]
data
```

```python
scat1 = Scatter(data, x='lt', y='mota', color='game')
show(scat1)
```

```python
data_lt = data.copy()
data_lt['loss'] = data_lt['lt']
data_lt = data_lt.groupby(['lt']).mean()
```

```python
scat2 = Scatter(data_lt, x = 'loss', y='mota')
scat3 = Scatter(data_lt, x = 'loss', y='motp')
scat4 = Scatter(data_lt, x = 'loss', y='mostly_tracked')
scat5 = Scatter(data_lt, x = 'loss', y='mt_auc')
show(bokeh.layouts.column(bokeh.layouts.row(scat2, scat3), bokeh.layouts.row(scat4, scat5)))
```

```python
mota = pd.DataFrame({'Score': data_lt['mota'], 'lost threshold': data_lt['loss'], 'Type': 'mota'})
motp = pd.DataFrame({'Score': data_lt['motp'], 'lost threshold': data_lt['loss'], 'Type': 'motp'})
motr = pd.DataFrame({'Score': data_lt['motr'], 'lost threshold': data_lt['loss'], 'Type': 'motr'})
mostly_tracked = pd.DataFrame({'Score': data_lt['mostly_tracked'], 'lost threshold': data_lt['loss'], 'Type': 'mostly_tracked'})
mt_auc = pd.DataFrame({'Score': data_lt['mt_auc'], 'lost threshold': data_lt['loss'], 'Type': 'mt_auc'})
frames = [mota, motp, motr, mostly_tracked, mt_auc]
data2 = pd.concat(frames)
data2
```

```python
scat11 = Scatter(data2, x='lost threshold', y='Score', color='Type')
scat11.circle([2,2,2,2,2],data2[data2['lost threshold']==2]['Score'], 
              fill_color='white', size=20, fill_alpha=0, line_color='black')
scat11.legend.location = "top_right"
show(scat11)
```
