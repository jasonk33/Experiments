```python
import pandas as pd
import numpy as np
from bokeh.charts import Scatter
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
```

```python
output_notebook()
```

```python
data = pd.read_csv('/Users/jason.katz/Downloads/LT-DET1.csv')
```

```python
data['lt'] = 1
for index, row in data.iterrows():
    data['lt'][index] = row['job_id'][-6:-3].replace("-", "")
```

```python
data['det'] = 1
for index, row in data.iterrows():
    data['det'][index] = row['job_id'][-2:]
```

```python
data['game'] = 'id'
for index, row in data.iterrows():
    data['game'][index] = row['job_id'][22:28]
```

```python
from bokeh.layouts import column, row
scat_ht_1 = Scatter(data, x='det', y='mota', color='game')
scat_lt_1 = Scatter(data, x='lt', y='mota', color='game')
show(row(scat_ht_1,scat_lt_1))
```

```python
data_det = data.copy()
data_det['Minimum Detections'] = data_det['det']
data_det = data_det.groupby(['det']).mean()
data_det
```

```python
data_lt = data.copy()
data_lt['lost'] = data_lt['lt']
data_lt = data_lt.groupby(['lt']).mean()
data_lt
```

```python
scat_det_2 = Scatter(data_det, x = 'Minimum Detections', y='mota')
scat_det_3 = Scatter(data_det, x = 'Minimum Detections', y='motp')
scat_det_4 = Scatter(data_det, x = 'Minimum Detections', y='mostly_tracked')
scat_det_5 = Scatter(data_det, x = 'Minimum Detections', y='mt_auc')
from bokeh.layouts import column, row
show(column(row(scat_det_2, scat_det_3), row(scat_det_4, scat_det_5)))
```

```python
scat_lt_2 = Scatter(data_lt, x = 'lost', y='mota')
scat_lt_3 = Scatter(data_lt, x = 'lost', y='motp')
scat_lt_4 = Scatter(data_lt, x = 'lost', y='mostly_tracked')
scat_lt_5 = Scatter(data_lt, x = 'lost', y='mt_auc')
from bokeh.layouts import column, row
show(column(row(scat_lt_2, scat_lt_3), row(scat_lt_4, scat_lt_5)))
```
